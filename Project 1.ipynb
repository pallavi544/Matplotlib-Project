## matplotlib

#### bar

```python
import matplotlib.pyplot as plt 
```

```python
data=df.groupby('Primary language')['Subscribers (millions)'].mean()
data
```

```python
y=data.values
x=data.index
plt.figure(figsize=(20,10))
plt.bar(x,y,color='pink',edgecolor='black')
plt.title('Youtube subscription data(Primary language Vs country)')
plt.xlabel('Primary language')
plt.ylabel('Subscribers (millions)')
plt.show()
```

```python
data=df.groupby('Category')['Subscribers (millions)'].mean()
data
```

```python
y=data.values
x=data.index
plt.figure(figsize=(20,15))
plt.bar(x,y,color='aquamarine',edgecolor='black')
plt.title('Youtube subscription data(Primary language Vs country)')
plt.xlabel('category')
plt.ylabel('Subscribers (millions)')
plt.show()
```

#### Histogram

```python
x=df['Subscribers (millions)']
plt.figure(figsize=(20,10))
plt.hist(x,color='skyblue',edgecolor='black')
plt.title('Youtube subscription data')
plt.xlabel('Subscribers (millions)')
plt.show()
```

#### Piechart

```python
d=df.groupby('Country')['Subscribers (millions)'].mean().reset_index()
d
```

```python
x=d['Subscribers (millions)']
mylabels=d['Country']
mycolors=['Purple','Yellow','red','green','hotpink','orange','grey','blue','skyblue','cyan','magenta','silver','aquamarine','goldenrod','coral','gold']
plt.figure(figsize=(20,10))
plt.pie(x,labels=mylabels,autopct='%1.1f%%',shadow=True,colors=mycolors)
plt.legend(title='Youtube subscription data',loc='upper left',bbox_to_anchor=(1.0,1))
plt.title('Youtube subscription data')
plt.xlabel('Subscribers (millions)')
plt.show()
```
